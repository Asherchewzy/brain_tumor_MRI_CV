# Predicting on images saved from google

```python
from fastai import *
from fastai.vision.all import *
# import matplotlib.pyplot as plt
# import matplotlib.image as mpimg
from PIL import Image
```

## IMPORTANT NOTE: the model was trained in a Linux os, because it is transfered to a windows os, the following block is required to redirect PosixPath to WindowsPath.

```python
import pathlib
temp = pathlib.PosixPath
pathlib.PosixPath = pathlib.WindowsPath

learn = load_learner('model.pkl')
```

## Example 1

```python
im1 = PILImage.create('google_images/image_1.jpg')
im1.thumbnail((192,192))
im1
```

```python
learn.predict(im1)
```

## Example 2

```python
im2 = PILImage.create('google_images/image_2.jpg')
im2.thumbnail((192,192))
im2
```

```python
#another output alternative

cat = ("no_tumor", "tumor")

def classify(im):
    pred, idx, probs = learn.predict(im)
    return dict(zip(cat, map(float,probs)))

classify(im1)
```

```python
#another output alternative

pred, idx, probs = learn.predict(im1)

print(f"{pred}, this looks like a tumor")
print(f"Probability it's not tumor is: {probs[0]:.2f}")
print(f"Probability it's a tumor is: {probs[1]:.2f}")
```
